```python
#hàm crawl thông tin thông qua id sản phẩm
from bs4 import BeautifulSoup
import requests

def get_product_info(product_id):
    url = f"https://tiki.vn/api/v2/products/{product_id}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        
        # Lấy thông tin categories
        categories = data.get("categories", {})
        category_info = {
            "id": categories.get("id"),
            "name": categories.get("name")
        } if isinstance(categories, dict) else None

        breadcrumbs = data.get("breadcrumbs", [])
        fourth_category_name = None
        if len(breadcrumbs) >= 4:
            fourth_category_name = breadcrumbs[3].get("name", None)

        # Lấy danh sách các large_url
        images = data.get("images", [])
        large_urls = [image.get("large_url") for image in images if "large_url" in image]

        # Lấy thông tin authors
        authors = data.get("authors", [])
        authors_info = [{"id": author.get("id"), "name": author.get("name")} for author in authors]

        # Xử lý description để loại bỏ thẻ HTML
        raw_description = data.get("description", "")
        soup = BeautifulSoup(raw_description, "html.parser")
        description = soup.get_text()  # Lấy văn bản thuần túy từ HTML

        # Lọc các thông tin cần thiết
        product_details = {
            "id": data.get("id"),
            "name": data.get("name"),
            "price": data.get("price"),
            "description": description,  # Description không có thẻ HTML
            "authors": authors_info,  # Danh sách các tác giả với id và name
            "quantity": data.get("review_count"),
            "image_urls": large_urls,
            "rating_average": data.get("rating_average"),
            "review_count": data.get("review_count"),
            "quantity_sold":data.get("all_time_quantity_sold"),
            "category": fourth_category_name 
            # category_info
        }
        return product_details
    else:
        print(f"Không thể lấy thông tin sản phẩm ID: {product_id}")
        return None
```

```python
import csv
import json
import time
import random


# Đọc danh sách product_id từ file CSV
def read_product_ids(file_path):
    product_ids = []
    with open(file_path, newline='', encoding='utf-8') as f:
        reader = csv.reader(f)
        for row in reader:
            product_ids.append(row[0])  # Giả sử ID sản phẩm là cột đầu tiên
    return product_ids

# Lưu thông tin sản phẩm vào file JSON
def save_product_info_to_json(product_info, output_file):
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(product_info, f, ensure_ascii=False, indent=4)

# Main process
def crawl_products(file_path, output_file):
    product_ids = read_product_ids(file_path)
    all_products_info = []

    for product_id in product_ids:
        print(f"Đang crawl sản phẩm ID: {product_id}")
        
        # Lấy thông tin sản phẩm
        product_info = get_product_info(product_id)
        if product_info:
            all_products_info.append(product_info)
        
        # Thời gian nghỉ ngẫu nhiên từ 1 đến 5 giây
        time.sleep(random.randint(1, 3))

    # Lưu tất cả thông tin sản phẩm vào file JSON
    save_product_info_to_json(all_products_info, output_file)
    print("Hoàn tất crawl thông tin sản phẩm!")

# Gọi hàm crawl sản phẩm
crawl_products("productId.csv", "products_info_1.json")
```
